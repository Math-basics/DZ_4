```python
# Загружаем библиотеки:
# для работы с математикой
import numpy as np
# для построения графиков
import matplotlib.pyplot as plt 
# для преобразования выражений
import sympy as sym
```

#### Задание 1

Задание: посчить значение первой и второй производной функции
$$    
f(x) = x^5 + 4sin(2x) + cos(3x+3)
$$

Первая производная
$$
f'(x) = (x^5)' + (4sin(2x))' + (cos(3x+3))' =\\
= 5x^4 + 4cos(2x)(2x)' - sin(3x+3)(3x+3)' =\\
= 5x^4 + 8cos(2x) - 3sin(3x+3)
$$

Вторая производная
$$
f''(x) = (5x^4)' + (8cos(2x))' - (3sin(3x+3))' =\\
= 20x^3 - 8sin(2x)(2x)' - 3cos(3x+3)(3x+3)' =\\
20x^3 - 16sin(2x) - 9cos(3x+3)
$$
В точке x = 1 значение -3,19

```python
x = 1
print(f'Значение первой производной в точке x = 1: {round((5*x**4 + 8*np.cos(2*x) - 3*np.sin(3*x + 3)), 3)}')
print(f'Значение второй производной в точке x = 1: {round((20*x**3 - 16*np.sin(2*x) - 9*np.cos(3*x + 3)), 3)}')
```

```python
# Объявляем символ "х"
x = sym.symbols('x')
# Исходная функция
f = x ** 5 + 4 * sym.sin(2 * x) + sym.cos(3 * x + 3)
# Первая производная
f_1 = sym.diff(f, x)
print(f'Первая производная {f_1}\nЗначение в точке x = 1: {f_1.evalf(3, subs={x: 1})}\n')
# Вторая производная
f_2 = sym.diff(f, x, 2)
print(f'Вторая производная {f_2}\nЗначение в точке x = 1: {f_2.evalf(3, subs={x: 1})}\n')
```

```python
# Объявляем символ "х"
x = sym.symbols('x')
# Исходная функция
f = sym.sin(2 * x + 1) ** 5
# Первая производная
f_1 = f.diff(x)
# Вторая производная
f_2 = f_1.diff(x)

# Строим графики на одном поле
func_plot = sym.plot(f, f_1, f_2, (x, -5, 5), size = (14, 7), xlabel = 'X', ylabel = '', legend = True, show = False, xlim = (-5, 5))
func_plot[0].line_color = 'r'
func_plot[1].line_color = 'g'
func_plot[2].line_color = 'b'
func_plot[0].label = 'Исходная функция'
func_plot[1].label = 'Первая производная'
func_plot[2].label = 'Вторая производная'
func_plot.show()
```
