```python
# Загружаем библиотеки:
# для работы с математикой
import numpy as np
# для расчёта производных
from scipy.misc import derivative
# для построения графиков
import matplotlib.pyplot as plt 
```

### Уровень 0

Задание: посчить значение первой и второй производной функции
$$    
f(x) = x^5 + 4sin(2x) + cos(3x+3)
$$

Первая производная
$$
f'(x) = 5x^4 + 8cos(2x) - 3sin(3x+3)
$$
Вторая производная
$$
f''(x) = 20x^3 - 16sin(2x) - 9cos(3x+3)
$$

```python
# Точка для расчёта значения фцнкций
x = 1
# Исходная функция
def f(x):
    return x ** 5 + 4 * np.sin(2 * x) + np.cos(3 * x + 3)
# Первая производная
def f_1(x):
    return 5 * x ** 4 + 8 * np.cos(2 * x) - 3 * np.sin(3 * x + 3)
# Вторая производная
def f_2(x):
    return 20 * x ** 3 - 16 * np.sin(2 * x) - 9 * np.cos(3 * x + 3)

print('В точке x = 1:')
print(f'- расчётное значение первой производной f(x)\' = {round(f_1(x), 3)}, scipy.misc.derivative() даёт {round(derivative(f, x, n = 1, dx = 1e-6), 3)}')
print(f'- расчётное значение второй производной f(x)\'\' = {round(f_2(x), 3)}, scipy.misc.derivative() даёт {round(derivative(f, x, n = 2, dx = 1e-6), 3)}')
```

```python
# Загружаем библиотеку для преобразования выражений
from sympy import *
from sympy.plotting import plot 
# Объявляем символ "х"
x = symbols('x')
# Исходная функция
f = sin(2 * x + 1) ** 5
# Первая производная
f_1 = f.diff(x)
# Вторая производная
f_2 = f_1.diff(x)

# Строим графики
plot(f, (x, -5, 5), line_color = 'red', title = 'Исходная функция', xlim = (-5, 5), ylabel = '')
plot(f_1, (x, -5, 5), line_color = 'blue', title = 'Первая производная', xlim = (-5, 5), ylabel = '')
plot(f_2, (x, -5, 5), line_color = 'green', title = 'Вторая производная', xlim = (-5, 5), ylabel = '')
plt.show() 
```
